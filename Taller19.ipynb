```python
import matplotlib
import math
import numpy
import matplotlib.pyplot as mpl
x=[1,2,3,4,5,6,7]
y=[0.2,0.5,1.8,3.4,5.7,9,13.8]

multixy=[x[i] * y[i] for i in range (len(x))]
xelevado2=[x[j]**2 for j in range (len(x))]
sumaxelevado2=sum(xelevado2)
sumamultixy=sum(multixy)
sumax=sum(x)
sumay=sum(y)
promx=sumax/(len(x))
promy=sumay/(len(x))
a1=(len(x)*(sumamultixy) - (sumax*sumay))/(len(x)*(sumaxelevado2) - ((sumax)**2))
a0=promy - (a1*promx)
print (f"Tu ecuación por regresiones lineales por minimos cuadrados para un mejor ajuste es: \n y={a0} + ({a1}x)\n")

st=[(y[i] - promy)**2 for i in range (len(x))]
sumst=sum(st)
sy=math.sqrt(sumst/(7-1))
sr=[(y[i] - a0 - (a1*x[i]))**2 for i in range (len(x))]
sumsr=sum(sr)
sy_x=math.sqrt(sumsr/(7-2))
r=math.sqrt((sumst-sumsr)/sumst) * 100
print(f"La desviacion estandar es: {sy} \nEl error estandar de la estimación es: {sy_x} \nEl coeficiente de correlación es: {r}")

#Grafica
graficax=[]
for i in range (int(x[0]),int(x[-1])+1):
  k=i
  while k<i+1:
    graficax.append(round(k,1))
    k+=0.1
graficay=[(a0+(a1*graficax[i])) for i in range (len(graficax))]

mpl.scatter(x,y, color= "green", marker="*", s=30)
mpl.plot(graficax, graficay, label="Lineal")
mpl.xlabel(x)
mpl.ylabel(y)
mpl.title("Grafica")
mpl.show()
```
